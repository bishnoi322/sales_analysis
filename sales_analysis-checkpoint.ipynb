```python
# Group the data by 'Product' and sum the 'Sales' for each product
product_sales = amazon.groupby('Category')['Amount'].sum().reset_index()
```

```python
# Sort the products by total sales in descending order
product_sales = product_sales.sort_values(by='Amount', ascending=False).head(6)
```

```python
# Plot the top-selling products
plt.figure(figsize=(10, 6))
plt.bar(product_sales['Category'], product_sales['Amount'])
plt.title('Top-Selling Products')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout() 
# Show the plot
plt.show()
```

```python
# Group the data by 'State' and sum the 'Amount' for each country
state_sales = amazon.groupby('ship-state')['Amount'].sum().reset_index()
```

```python
# Sort the states by total sales in descending order
state_sales = state_sales.sort_values(by='Amount', ascending=False).head(6)
```

```python
# Plot the states with the highest sales
plt.figure(figsize=(10, 6))
colors = ['red' if c == 'TELANGANA' else 'blue' for c in state_sales['ship-state']]  # USA in blue, others in green
plt.bar(state_sales['ship-state'], state_sales['Amount'], color=colors)
plt.title('Countries with the Highest Sales')
plt.xlabel('ship-state')
plt.ylabel('Total Sales')
plt.tight_layout()

# Show the plot
plt.show()
```
